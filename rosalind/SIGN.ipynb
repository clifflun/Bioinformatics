```python
import itertools
```

```python
def main(n):
    tmp=[]
    for i in range(-n,n+1):
        if i == 0:
            continue
        tmp.append(i)
    test = set([abs(i) for i in tmp])
    count=0
    for i in itertools.combinations(tmp,n):
        if set([abs(a) for a in i]) == test:
            for j in itertools.permutations(i):
                for k in j: 
                    print(k, end = ' ')
                print()
                count+=1
    return count
    
```

```python
main(4)
```
