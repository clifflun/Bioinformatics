This solution uses an initialized array and loop through the array only ***ONCE***.

The time complexity is technically O(n^2). Since the algorithm traces back for the longest match when the previous position is a match, but the current position is ***NOT*** an extension of it.

Best case is O(n).

In reality, it is closer to O(n) than O(n^2). (Unless there are lots of repeats in the prefix.)

Not the most elegant code, but it works.

Edge cases to consider: short tandem repeats (CAGCAGCAG), single nucleotide repeats (AAAAAACAAAAA)

```python
##getting fasta in list format
fa_id = []
temp = ''
for line in open('KMP.txt'): 
    if line.startswith('>'):
        fa_id.append(line.strip()[1:])
        temp +='>'
    else: 
        temp += line.strip()
fa_seq = temp.split('>')[1:]

print(fa_id)
# print(fa_seq)
```

```python
s = fa_seq[0]
def failure(s):
    tmp = [0]*len(s) ##initialize array with zeros
    for i in range(1,len(s)):
        if tmp[i-1] != 0: ## if previous position is a match
            if s[0:tmp[i-1]+1] == s[i-tmp[i-1]:i+1]:   ## if increasing length is a match        
                tmp[i] = tmp[i-1] + 1  ## previous length + 1
                continue
            else:    ## finding next longest possible prefix (for reapeating elements)
                j = tmp[i-1]
                while j >0:
                    if s[0:j] != s[i-j+1:i+1]: ##decreasing in matching length to find longest match
                        j-=1
                    else:
                        tmp[i] = j ## j == prefix match length
                        break
                continue                      
        if s[0] == s[i]: # if match first char, assign 1
            tmp[i] = 1
    return tmp
```

```python
%%time 
# s= 'AAACAAAAACATAATAA'
out = failure(s)
out = [str(o) for o in out]
```
