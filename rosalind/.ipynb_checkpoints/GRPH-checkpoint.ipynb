```python
##getting fasta in list format
fa_id = []
temp = ''
for line in open('GRPH.txt'): 
    if line.startswith('>'):
        fa_id.append(line.strip()[1:])
        temp +='>'
    else: 
        temp += line.strip()
fa_seq = temp.split('>')[1:]
```

```python
prefix = [i[:3] for i in fa_seq]
postfix = [i[-3:] for i in fa_seq]
```

```python
for i in range(len(fa_seq)): 
    for j in range(len(fa_seq)):
        if i == j: 
            pass
        elif fa_seq[i] == fa_seq[j]: 
            pass
        else:
#             print(i, j)
            if postfix[i] == prefix[j]:
                print(f"{fa_id[i]} {fa_id[j]}")
```
