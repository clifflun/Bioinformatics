```python
import itertools
import LEXF
```

```python
##getting fasta in list format
fa_id = []
temp = ''
for line in open('KMER.txt'): 
    if line.startswith('>'):
        fa_id.append(line.strip()[1:])
        temp +='>'
    else: 
        temp += line.strip()
fa_seq = temp.split('>')[1:]

print(fa_id)
print(fa_seq)
```

```python
def main(seq, k):
    nu=['A','C','G','T']
    nuk = LEXF.main(nu,k)
    kcomp = {}
    for n in nuk:
        kcomp[n] = 0
    for i in range(0,len(seq)-k+1):
        tmp=seq[i:i+k]
        kcomp[tmp] = kcomp.get(tmp,0) + 1
    kcount = [kcomp[i] for i in sorted(kcomp.keys())]
    for k in kcount:
        print(k, end = ' ')
```

```python
if __name__=="__main__":
    seq = fa_seq[0]

    main(seq,4)
```
