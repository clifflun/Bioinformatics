```python
##getting fasta in list format
fa_id = []
temp = ''
for line in open('CONS.txt'): 
    if line.startswith('>'):
        fa_id.append(line.strip()[1:])
        temp +='>'
    else: 
        temp += line.strip()
fa_seq = temp.split('>')[1:]
```

```python
##getting list of dicts of count at each position
master_count = []
for i in range(len(fa_seq[0])):
    count_dic = {}
    for j in range(len(fa_seq)):
        aa = (fa_seq[j][i])
        count_dic[aa] = count_dic.get(aa, 0)+1
    master_count.append(count_dic)
```

```python
## output formatting
con = ''
A = ''
C = ''
G = ''
T = ''
for pos in master_count:
    con += (max(pos, key = pos.get))
    A += str(pos.get('A', 0)) + ' '
    C += str(pos.get('C', 0)) + ' '
    G += str(pos.get('G', 0)) + ' '
    T += str(pos.get('T', 0)) + ' '
```

```python
print(con)
print(f'A: {A}')
print(f'C: {C}')
print(f'G: {G}')
print(f'T: {T}')
```
